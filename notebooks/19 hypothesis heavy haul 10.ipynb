# 10: Hva har total forbruk og gj. snitt forbruk vært for ulike ekvipasjer

```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import matplotlib.pyplot as plt
from source.features_dir import vehicle_data_74t
```

```python
df = vehicle_data_74t.df_truck_vehicle_data()
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df['mil'] = df['Distanse (km)'] / 10

df_agg = df.groupby(['ekvipasje']).agg(
    liter_kjoring=('Liter kjøring', 'sum'),
    mil=('mil', 'sum')
).reset_index()

df_agg['forbruk'] = df_agg['liter_kjoring'] / df_agg['mil']
df_agg
```

```python
df_agg = df.groupby(['år', 'ekvipasje']).agg(
    liter_kjoring=('Liter kjøring', 'sum'),
    mil=('mil', 'sum')
)

df_agg['forbruk'] = df_agg['liter_kjoring'] / df_agg['mil']
df_agg
```
