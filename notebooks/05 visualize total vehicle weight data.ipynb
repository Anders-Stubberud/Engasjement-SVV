```python
from source import config
import os

# Define the directory
image_dir = config.VEHICLE_WEIGHT_FIGURES_DIR_WIM

# List all files in the directory
images = [
    os.path.join(image_dir, file)  # Construct the full file path
    for file in os.listdir(image_dir)
    if any(file.endswith(ext) for ext in ['.png'])
]

# Optionally, sort the files by different criteria
images.sort()  # Sort alphabetically by path (filename)

# Print out the sorted image paths
for img in images:
    print(f'"{str(img).split("/")[-1].strip(".png")}"')
```
